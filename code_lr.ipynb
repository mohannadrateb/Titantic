Import the data and read it

```python
import pandas as pd
data = pd.read_csv("/content/train.csv")
```

see the prdictors and values 

```python
data.head()
```

drop non needed columns 

```python
data = data.drop(columns =['Name','Ticket']) 
```

Get the different types of data

```python
data.describe()
```

```python
data.head()
```

```python
data[["Sex","Embarked"]] = data[["Sex","Embarked"]].astype('category')
data["Sex"] = data["Sex"].cat.codes
data["Embarked"] = data["Embarked"].cat.codes
```

```python
data.head()
```

```python
data = data.drop(columns = ["Cabin"])
```

```python
y = data["Survived"]
data = data.drop(columns= ["Survived"])
```

```python
import pandas as pd
test = pd.read_csv("/content/test.csv")
test = test.drop(columns =['Name','Ticket','Cabin']) 
test[["Sex","Embarked"]] = test[["Sex","Embarked"]].astype('category')
test["Sex"] = test["Sex"].cat.codes
test["Embarked"] = test["Embarked"].cat.codes
test = test.fillna(test.mean())
```

make sure to replace the nan values  

```python
data = data.fillna(data.mean())
```

```python
y.isnull().values.any()
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
#train the lr model
lr.fit(data,y)
#print("acc of the log reg model is:{}".format(lr.score()))
```

```python
data.head()
```

```python
test.head()
```

```python
# predict for the test data
predictions_survived =lr.predict(test)
```

```python
dic_data = {'PassengerId': test["PassengerId"]
     ,'Survived': predictions_survived
     }
submissions = pd.DataFrame(data= dic_data,index =False)
```
